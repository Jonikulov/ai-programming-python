# Histograms

```python
# importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# load the dataset
pokemon = pd.read_csv("data/pokemon.csv")
pokemon.head()
```

## `matplotlib.pyplot.hist()`

Syntax: `matplotlib.pyplot.hist(x, bins=None, range=None, density=False, *, data=None, **kwargs)`

Compute and plot a histogram.

### Plot a default histogram

```python
# Plot the speed column on histogram
plt.hist(data=pokemon, x='speed', edgecolor='black')
# We have intentionally not put a semicolon at the end of the statement below to see the bar-width
```

### Histogram with fixed number of bins

```python
plt.hist(data=pokemon, x='speed', bins=20, ec='black')
```

### Histogram with dynamic number of bins

```python
# Create bins with step-size 5
bins = np.arange(0, pokemon['speed'].max()+5, 5)
plt.hist(data=pokemon, x='speed', bins=bins, ec='black')
```

## `seaborn.displot()`

Syntax: `seaborn.displot(data=None, x=None, y=None, ..., kind='hist', kde=False)`

This function can also plot histograms, as similar to the `pyplot.hist()` function, and is integrated with other univariate plotting functions.

```python
sns.displot(data=pokemon, x='speed')
```

```python
bin_edges = np.arange(0, pokemon['speed'].max()+1/2, 1/2)
sns.displot(data=pokemon, x='speed', bins=bin_edges, alpha=1)
```

```python
bin_edges = np.arange(0, pokemon['speed'].max()+1/4, 1/4)
sns.displot(data=pokemon, x='speed', bins=bin_edges, alpha=1)
```

### `matplotlib.pyplot.subplot`

Syntax: `subplot(nrows, ncols, index, **kwargs)`

Add an Axes to the current figure or retrieve an existing Axes.

### Plot two histograms side-by-side

```python
# Set a larger figure size for subplots
plt.figure(figsize=(20, 5))

# histogram on left, example of too-large bin size
plt.subplot(1, 2, 1)  # 1 row, 2 cols, subplot 1
bins = np.arange(0, pokemon['speed'].max()+4, 4)
plt.hist(data=pokemon, x='speed', bins=bins)

# histogram on right, example of too-small bin size
plt.subplot(1, 2, 2)  # 1 row, 2 cols, subplot 2
bins = np.arange(0, pokemon['speed'].max()+1/4, 1/4)
plt.hist(data=pokemon, x='speed', bins=bins, alpha=1)

plt.show()
```
